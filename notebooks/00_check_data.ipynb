```python
import yfinance as yf
import pandas as pd

# 1) QQQ 일봉 데이터 다운로드
df = yf.download("QQQ", start="2005-01-01", interval="1d")

# 2) 기본 정보 확인
print("Shape:", df.shape)
display(df.head())
display(df.tail())

# 3) 결측치 개수 확인
print("\nMissing values per column:")
display(df.isna().sum())
```

```python
df_bd = df.asfreq('B')  # Business day frequency
df_filled = df_bd.ffill()  # 직전 거래일 종가·값으로 채움
print("Before fill:", df_bd.isna().sum().to_dict())
print("After  fill:", df_filled.isna().sum().to_dict())
# df_bd['Close'].plot(title="QQQ Close (with gaps before fill)")
# df_bd 는 영업일 빈도로 다시 샘플링한 데이터프레임
# df_filled['Close'].plot(title="QQQ Close (post-fill)")
# df_bd의 결측값을 ffill()로 채워넣은 데이터프레임
# 결측값 : 데이터 셋에서 해당 시점이나 관측치에 실제 값이 기록되지 않아 비어있는 상태
```
